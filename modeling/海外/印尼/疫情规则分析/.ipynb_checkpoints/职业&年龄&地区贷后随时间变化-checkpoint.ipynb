# 函数

```python
def data_process(df,group_var):
    df_temp = df.copy()
    table_temp = df_temp.groupby(group_var).agg({'total_loan':'sum', 
                                             'dpd1':'sum',
                                             'dpd3':'sum',
                                             'dpd7':'sum'}).reset_index()
    table_temp['dpd1+'] = table_temp['dpd1']/table_temp['total_loan']
    table_temp['dpd3+'] = table_temp['dpd3']/table_temp['total_loan']
    table_temp['dpd7+'] = table_temp['dpd7']/table_temp['total_loan']
    table_temp['cuihui13'] = (table_temp['dpd1+']-table_temp['dpd3+'])/table_temp['dpd1+']
    table_temp['cuihui17'] = (table_temp['dpd1+']-table_temp['dpd7+'])/table_temp['dpd1+']
    return table_temp
# data_process(df,'due_date')

def buchang(df,var,n): 
    df_temp = df.copy()
    a = df_temp[var].unique()
    step = n
    b = [a[i:i+step] for i in range(0,len(a),step)]
    my_dict = {}
    for index,values in enumerate(b):
        my_dict[values[0]] = values
    df_temp['date'] = df_temp[var].apply(lambda x: [k for k,v in my_dict.items() if x in v][0])
    return df_temp
# buchang(df,'due_date',3).head()

def mylist(df, var):
    df_temp = df.copy()
    if var == 'industry_involved':
        table1 = pd.pivot_table(df_temp, index=var, values='total_loan',
                        aggfunc='sum').reset_index().sort_values(by='total_loan', ascending=False)
        index_list1 = table1.head(10)[var].tolist()
        index_list2 = table1[10:][var].tolist()
        index_list3 = table1[var].tolist()
    if  var  in ['province','city',	'county']:
        table1 = pd.pivot_table(df_temp,index = var,values='total_loan',aggfunc='sum').reset_index().sort_values(by='total_loan',ascending=False)
        table1['rate'] = table1['total_loan']/table1['total_loan'].sum()
        table1 = table1[table1['rate']>0.01]
        index_list1 = table1.head(10)[var].tolist()
        index_list2 = table1[10:][var].tolist()
        index_list3 = table1[var].tolist()
    return var,index_list1,index_list2,index_list3
        
```
