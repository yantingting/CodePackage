# 数据合并

```python
df_all1 = load_data_from_pickle(file_path,'df_all1.pkl')
drop_list =  ['marital_status','religion','education','media_source','provincecode',\
              'citycode','districtcode','gender','brand','last_apply_now','last_paid_now']
df_all1.drop(drop_list,axis =1,inplace = True)
df_all1 = df_all1[df_all1['is_new']==0]
df_all1.shape 
df_all1.head()
var_app = load_data_from_pickle(file_path,'var_app.pkl')
var_app.shape
var_app.head()
```

```python
drop_list1 = ['customer_id','reg_date','is_new','dpd7+','profit_rmb','freq']
var_dict_base = dp.VarDict(df_all1.drop(drop_list1, axis = 1), data_sorce='baseinfo')
var_dict_base.to_csv(os.path.join(file_path, 'var_dict_base.csv'), index = False)
var_dict_app = dp.VarDict(var_app, data_sorce='app')
var_dict_app.to_csv(os.path.join(file_path, 'var_dict_app.csv'), index = False)
print('success')
```

```python
time1 = time.time()
var_all = df_all1.merge(var_app,left_on='customer_id',right_index=True,how='left')
eda_result = ss.eda(var_all,file_path)
var_all.fillna(-1,inplace = True)
save_data_to_pickle(var_all,file_path,'var_all.pkl')
var_all.shape
var_all.head()
time2 = time.time()
print('run_time: ', time2 - time1)
```
