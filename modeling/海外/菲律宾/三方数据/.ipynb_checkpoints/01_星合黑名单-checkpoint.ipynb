# 2)抽取样本标签

```python
query = '''
select distinct apply.order_no as order_no
 , apply.id_type as id_type
 , md5(apply.id_num) as id_num
 , md5(apply.customer_phone_num) as customer_phone_num
 , apply.created_at AS created_at
 , 'black' as test_type
 , 0 as latedays
 ,date('1970-01-01') as effective_date
from public.dw_flb_approval_phl_approval_prod_approval_applications as apply 
left join public.dw_flb_approval_phl_approval_prod_application_reasons as reasons 
on apply.order_no = reasons.order_no
where apply.product_name in ('Peso2Go','JetPeso')
 and reasons.reason_type = 'reject' and (reasons.reason_code = 'R204'  or reasons.reason_code = 'R202') and date(apply.created_at)<'2019-12-10'
 UNION ALL 
  select distinct apply.order_no as order_no
 , apply.id_type as id_type
 , md5(apply.id_num) as id_num
 , md5(apply.customer_phone_num) as customer_phone_num
 , apply.created_at AS created_at
 , 'rule' as test_type
 , 0 as latedays
  ,date('1970-01-01') as effective_date
from public.dw_flb_approval_phl_approval_prod_approval_applications as apply 
left join public.dw_flb_approval_phl_approval_prod_application_reasons as reasons 
on apply.order_no = reasons.order_no
where apply.product_name in ('Peso2Go','JetPeso')
and reasons.reason_type = 'reject' and reasons.reason_code != 'R204' and reasons.reason_code != 'R202' and date(apply.created_at)>'2019-11-30' and  date(apply.created_at)<'2019-12-10'
UNION ALL 
select t1.order_no as order_no
, t3.id_type as id_type
, t3.id_num AS id_num
, t3.customer_phone_num AS customer_phone_num
, t3.created_at AS created_at
, 'fund' as test_type
, t2.DPD as latedays
,t1.effective_date as effective_date
from 
(select order_no,effective_date  from public.dw_flb_hx_phl_core_prod_core_loans 
where order_type in ('NewApplicationOrder' , 'RepeatApplicationOrder') and product_name in ('Peso2Go','JetPeso')
and effective_date between '2019-08-01' and '2019-11-20') t1 
left join 
(SELECT order_no,
case when actual_paid_off_date = '1970-01-01' then  '2019-12-22' - due_date
else actual_paid_off_date-due_date end AS DPD
FROM public.dw_flb_hx_phl_core_prod_core_repayment_schedules
WHERE product_name in ('Peso2Go','JetPeso')
and  extend_period = 0) t2 
on t1.order_no = t2.order_no
left join 
(select order_no
, id_type as id_type
 , md5(customer_full_name) as customer_full_name
 , md5(id_num) as id_num
 , md5(customer_phone_num) as customer_phone_num
 , created_at
from public.dw_flb_approval_phl_approval_prod_approval_applications) t3
on t1.order_no = t3.order_no;
'''
```

```python
df_target = data_con.database_gn(query)
df_target.info()
df_target['temp'] = df_target.apply(lambda x: ''.join([x['id_num'],str(x['created_at'])]), axis=1)
df_target['apply_date'] = df_target['created_at'].apply(lambda x: str(x)[:10])
df_target['dpd7+'] = df_target['latedays'].apply(lambda x: 1 if x>=7 else 0)
df_target['temp'].nunique()
df_target.to_csv(file_path + 'df_target.csv', index = False)
```

```python
my_df = pd.read_excel(file_path + 'my_df.xlsx')
```

```python
my_df_fund = my_df[my_df['test_type'] == 'fund']
my_df_fund.shape
```

```python
var_list = ['共用手机号','共用身份证','灰名单','黑名单']
for col in var_list:
    print('{} is in processing'.format(col))
    frame1 = dp.ContinuousBin(my_df_fund, col , target = 'dpd7+', n = 10, special_attribute = [],method = 'equal freq')[0]
    ks = dp.ContinuousBin(my_df_fund, col , target = 'dpd7+', n = 10, special_attribute = [],method = 'equal freq')[1]
    iv = dp.ContinuousBin(my_df_fund, col , target = 'dpd7+', n = 10, special_attribute = [],method = 'equal freq')[2]
    frame1
    print('KS = ', ks)
    print('IV = ', iv)
    
```

```python
my_df_black = my_df[my_df['test_type'] == 'black']
my_df_black.shape
```

```python
var_list = ['共用手机号','共用身份证','灰名单','黑名单']
total = my_df_black.shape[0]
for col in var_list:
    print('{} is in processing'.format(col))
    d = Counter(my_df_black[col])
    temp_frame = pd.DataFrame.from_dict(d, orient='index').reset_index().rename(columns = {'index':col,0: 'N'})
    temp_frame['rate'] = round(temp_frame['N']/total,3)
    temp_frame
    
```
