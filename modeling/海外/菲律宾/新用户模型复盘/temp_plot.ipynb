```python
x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x)
#   配置
plt.figure(figsize=(8,4))       #   框的大小
plt.plot(x,y,label='$sin(x)$')
plt.plot(x,z,'b--',label='$cos(x)$')        #   蓝色虚线
plt.xlabel("Time(s)")
plt.ylabel("")
plt.title("您好！")
plt.ylim(-1.2, 1.2)
plt.legend()                        #   显示为一个样式
plt.show();

fig = plt.gcf()                         #   获得当前图标的对象
ax = plt.gca()                      #   获得子图的对象
print(fig)
print(ax)
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot();
# ts
```

```python
plt.figure()
ts.plot(style='k--', label='Series')
plt.legend();
```

```python
df =pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD')) 
df = df.cumsum()
plt.figure()
df.plot();
plt.legend(loc='best');
```

```python
df.plot(subplots=True, figsize=(6, 6))
plt.legend(loc='best');
```

```python

plt.figure();  
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))  
ts = np.exp(ts.cumsum())  
ts.plot(logy=True) #对y轴进行log(y)放缩，图中y轴刻度依然是y的真实值，而不是log(y)  
plt.title(u'画图');
```
