# 巴西银行数据
不是所有的样本都用来进行分析，取消的订单不作为样本

```python
bank_data = pd.read_excel(file_path + '巴西字典.xlsx', sheet_name = 'PORTOCRED DATABASE',na_values='',dtype=({'ZIP CODE': str}))
bank_data = bank_data.rename(columns = {'LOAN ID': 'loan_id'})
print(bank_data.shape)
drop_list= ['PROPOSTA CANCELADA',  'VALIDADE EXPIRADA', 'SIMULAÇÃO']  #剔除无效样本
bank_data = bank_data[~bank_data['LOAN CONTRACT STATUS'].isin(drop_list)]
reject_list = ['REPROVADO']
bank_data['flag_result'] = bank_data['LOAN CONTRACT STATUS'].apply(lambda x: 0 if x in reject_list else 1)
#生成银行数据的所有loan
bank_data[['loan_id', 'DATE', 'flag_result']].to_csv(os.path.join(file_path_out, 'loan_bank.csv'), index = False)
print(bank_data.shape)
bank_data.head(2)
```

银行数据生成变量（未处理）

```python
remain_col = ['loan_id', 'DATE','flag_result' , 'SERASA SCORE','BOA VISTA (BVS) SCORE']
var_bank_data = bank_data[remain_col].copy()
var_bank_data['age'] = bank_data.apply(lambda x: int(str(x['DATE'])[:4]) - pd.to_datetime(x['BIRTH DATE (CLIENT INFORM)']).year, axis = 1)
marital_dict = dict(zip(bank_data['MARITIAL STATUS'].unique().tolist(), range(len(bank_data['MARITIAL STATUS'].unique()))))
var_bank_data['marital'] = bank_data['MARITIAL STATUS'].map(lambda x: marital_dict[x])
var_bank_data['len_zipcode'] = bank_data['ZIP CODE'].apply(lambda x: len(x) if not isinstance(x, float) else 0)
occupation_dict = dict(zip(bank_data['OCCUPATION'].unique().tolist(), range(len(bank_data['OCCUPATION'].unique()))))
var_bank_data['occupation'] = bank_data['OCCUPATION'].map(lambda x: occupation_dict[x])
gender_dict = dict(zip(bank_data['GENDER'].unique().tolist(), range(len(bank_data['GENDER'].unique()))))
var_bank_data['gender'] = bank_data['GENDER'].map(lambda x: gender_dict[x])
var_bank_data['EMAIL'] = bank_data['EMAIL'].apply(lambda x: x[x.find('@') + 1:x.find('.')] if not isinstance(x, float) else 0)
email_dict = dict(zip(var_bank_data['EMAIL'].unique().tolist(), range(len(var_bank_data['EMAIL'].unique()))))
var_bank_data['email'] = var_bank_data['EMAIL'].map(lambda x: email_dict[x])
var_bank_data['nationality'] = bank_data['NATIONALITY']
flag0_dict = dict(zip(bank_data['NOTARYS OFFICE PROTEST FLAG'].unique().tolist(), range(len(bank_data['NOTARYS OFFICE PROTEST FLAG'].unique()))))
var_bank_data['flag0'] = bank_data['NOTARYS OFFICE PROTEST FLAG'].map(lambda x: flag0_dict[x])

flag1_dict = dict(zip(bank_data['REATIL DEBT DELAY FLAG'].unique().tolist(), range(len(bank_data['REATIL DEBT DELAY FLAG'].unique()))))
var_bank_data['flag1'] = bank_data['REATIL DEBT DELAY FLAG'].map(lambda x: flag1_dict[x])
flag2_dict = dict(zip(bank_data['BANK DEBT DELAY FLAG'].unique().tolist(), range(len(bank_data['BANK DEBT DELAY FLAG'].unique()))))
var_bank_data['flag2'] = bank_data['BANK DEBT DELAY FLAG'].map(lambda x: flag2_dict[x])
flag3_dict = dict(zip(bank_data['LAWSUIT FLAG'].unique().tolist(), range(len(bank_data['LAWSUIT FLAG'].unique()))))
var_bank_data['flag3'] = bank_data['LAWSUIT FLAG'].map(lambda x: flag3_dict[x])
flag4_dict = dict(zip(bank_data['BOUNCED CHECK FLAG'].unique().tolist(), range(len(bank_data['BOUNCED CHECK FLAG'].unique()))))
var_bank_data['flag4'] = bank_data['BOUNCED CHECK FLAG'].map(lambda x: flag4_dict[x])

var_bank_data['flag6'] = bank_data['ACTIVE COMPANIES ON HIS NAME']
var_bank_data['flag7'] = bank_data['SOCIETARY PARTICIPATION STATUS'].apply(lambda x: 0 if isinstance(x,float) else 1)
flag8_dict = dict(zip(bank_data['CDL NEGATIVE FLAG'].unique().tolist(), range(len(bank_data['CDL NEGATIVE FLAG'].unique()))))
var_bank_data['flag8'] = bank_data['CDL NEGATIVE FLAG'].map(lambda x: flag8_dict[x])
var_bank_data['city'] = preprocessing.LabelEncoder().fit_transform(bank_data['CITY NAME'].fillna('missing'))
var_bank_data['bank_name'] = preprocessing.LabelEncoder().fit_transform(bank_data['BANK NAME'].fillna('missing'))
drop_list = ['loan_id', 'DATE', 'flag_result', 'EMAIL']
var_dict_bank = dp.VarDict(var_bank_data.drop(drop_list, axis = 1), data_sorce='bank')
var_dict_bank.to_csv(os.path.join(file_path_out, 'var_dict_bank.csv'), index = False)
var_bank_data.drop(['EMAIL','DATE', 'flag_result'],axis =1).to_csv(os.path.join(file_path_out, 'var_bank_data.csv'), index = False)
var_bank_data.shape
```
